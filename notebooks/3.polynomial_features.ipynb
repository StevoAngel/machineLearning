```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
m = 100
X = 6 * np.random.rand(m,1) - 3
y = 0.5 * X**2 + X + 2 + np.random.rand(m, 1)

plt.scatter(X, y)
plt.show()
```

```python
# Data augmentation:
poly = PolynomialFeatures(degree=2)

X_ext = poly.fit_transform(X)
```

```python
# Linear regression:
lin_reg = LinearRegression()
lin_reg.fit(X_ext,y) # Training model with X data


X_test = np.linspace(-3, 3, 100) # Linear space to test
X_test = X_test.reshape(-1, 1)
X_test_ext = poly.transform(X_test)

y_pred = lin_reg.predict(X_test_ext)

plt.scatter(X, y, color='blue', label='Real Data')
plt.plot(X_test, y_pred, color='red', label='Linear-Regression Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Reg')
plt.legend()
plt.show()
```

```python
# SGDRegressor:
sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, eta0=0.01, learning_rate='invscaling', random_state=42)
sgd_reg.fit(X_ext, y)

y_pred_sdg = sgd_reg.predict(X_test_ext)

plt.scatter(X, y, color='blue', label='Real Data')
plt.plot(X_test, y_pred_sdg, color='red', label='SGDRegressor Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Reg')
plt.legend()
plt.show()
```
