# Intro GD ITESO

$f(x) = x^2 + 1$

$f'(x) = 2x$

```python
initial_x = 5
alpha = 0.1
n_iterations = 50

iterations = list()

y = list() # Resultados

x = initial_x # Valor inicial

for i in range(n_iterations):
    # Calculate gradient for current x
    g = 2 * x

    # Update X with grafient
    x = x - alpha * g

    #Save the current value of f(x):
    y.append(x**2 + 1)

    iterations.append(i + 1)

    print(f"i: {i+1}, x = {x}, y/f(x) = {x**2 +1}") 
```

```python
import matplotlib.pyplot as plt
from numpy import linspace

plt.subplot(1, 2, 1)
plt.plot(iterations, y)
plt.xlabel("Iteration/Epochs")
plt.ylabel("Loss/y/f(x)/J/C")

plt.subplot(1, 2, 2)
X = linspace(-5, 5, 100)
Y = X**2 + 1
plt.plot(X, Y, x, y[-1], "ro")
```
